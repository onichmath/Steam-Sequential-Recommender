```python
# Scrape steam charts 
import requests 
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
```

```python
url_prefix = "https://steamcharts.com/app/" 
url_suffix = "#All"
```

```python
items = pd.read_parquet("./data/products_X.parquet") 
```

```python
products = items["product_id"].to_numpy()
print(products)
```

```python
def get_player_count(product_id):
    url = url_prefix + str(product_id) + url_suffix
    response = requests.get(url)
    
    if response.status_code != 200:
        print(f"Failed to get player count for {product_id}")
        return
    soup = BeautifulSoup(response.text, 'html.parser')
    player_count = soup.find("table", class_="common-table") 
    print(player_count)
```

```python
get_player_count(products[0])
```
